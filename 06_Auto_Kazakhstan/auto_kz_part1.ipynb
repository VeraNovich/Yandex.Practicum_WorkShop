### **0.2. Общие понятия автомобильного рынка**

**Классификация автомобилей по размеру:**

- **A-класс - «Особо малый класс»:** длина до 3,8 м, ширина — до 1,6 м.
- **B-класс - «Малый класс»:** длина — 3,8–4,5 м, ширина — 1,5–1,7 м.
- **C-класс - «Малый средний класс»:** длина — 4,2–4,7 м, ширина — 1,75–1,83 м.
- **D-класс - «Средний класс»:** длина — 4,6–4,9 м, ширина — 1,8–1,89 м.
- **E-класс - «Высший средний класс»:** длина — 4,8–5,0 м, ширина — более 1,8 м.
- **F-класс - «Представительский класс»:** длина — более 5,0 м, ширина — свыше 1,8 м.

**Классификация автомобилей по типу кузова:**

- **Кроссоверы:** автомобили повышенной проходимости.
- **Внедорожники:** автомобили, созданные для езды по бездорожью.
- **Пикапы:** грузопассажирские автомобили с минимальными грузовыми возможностями и комфортом легкового автомобиля.
- **Минивэны:** семейные автомобили повышенной вместимости.
- **Седаны:** четырёхдверные модели автомобилей, где багажное отделение структурно отделено от пассажирского салона.
- **Хэтчбеки:** компромисс между седаном и универсалом: короче седана, но имеет большие возможности багажного отделения.
- **Универсалы:** грузопассажирские варианты исполнения легковых автомобилей.

**Классификация автомобилей по типу привода:**

- **RWD** – задний привод,
- **FWD** – передний привод,
- **4WD** – полный привод,
- **2WD** – подключаемый

**Классификация автомобилей по типу коробки передач:**

- **МКПП** - механическая коробка переключения передач,
- **АКПП** - автоматическа коробка переключения передач,
- **РКПП** -роботизированаая коробка переключения передач,
- **CVT** - вариаторная коробка переключения передач.

**Классификация автомобилей по типу двигателя:**

- бензиновый;
- газовый;
- водородный;
- дизельный;
- электромобиль;
- гибридный автомобиль.

**Автомобильные транспортные средства (АТС):**

- пассажирские (легковые автомобили и автобусы);
- грузовые;
- специального назначения.

В каждой из стран классификация автомобилей происходила с учётом национальных особенностей.

#### **2.4.2. объем_двигателя_класс**

- особо-малый - до 1.2 литра;
- малый - 1.2 - 1.8 литра;
- средний - 1.8 - 3.5 литра;
- большой - свыше 3.5 литра

```python
# Объявим функцию engine_size, определяющую класс автомобиля по объему двигателя в л.
def engine_size(df0):
    if df0['объём_двигателя(л)']<1.3:
      return 'особо-малый'
    elif df0['объём_двигателя(л)']<1.9:
      return 'малый'
    elif df0['объём_двигателя(л)']<3.6:
      return 'средний'
    else:
      return 'большой'
```

```python
# Создадим столбец 'объём_двигателя_класс'
df0['объём_двигателя_класс'] = df0.apply(engine_size, axis = 1)
```

```python
# Проверим значения нового столбца
df0['объём_двигателя_класс'].value_counts()
```

#### **2.4.3. категория_заказа_количество**

```python
# Изучим столбец 'количество'
df0['количество'].value_counts()
```

```python
# Объявим функцию order_size_cat, определяющую категорию заказа по количеству заказанных автомобилей
def order_size_cat(df0):
    if df0['количество']==1:
      return '1 автомобиль'
    elif df0['количество']==2:
      return '2 автомобиля'
    elif df0['количество']<11:
      return '3-10 автомобилей'
    elif df0['количество']<21:
      return '11-20 автомобилей'
    else:
      return 'более 20 автомобилей'
```

```python
# Создадим столбец 'категория_заказа_количество'
df0['категория_заказа_количество'] = df0.apply(order_size_cat, axis = 1)
```

#### **2.4.4. категория_цены**

```python
# Объявим функцию price_cat, определяющую категорию заказа по цене автомобиля
def price_cat(df0):
    if df0['цена(usd)']<15500:
      return 'до 15.500 USD'
    elif df0['цена(usd)']<25000:
      return 'от 15.500 до 25.000 USD'
    elif df0['цена(usd)']<35000:
      return 'от 25.000 до 35.000 USD'
    elif df0['цена(usd)']<50000:
      return 'от 35.000 до 50.000 USD'
    else:
      return 'более 50.000 USD'
```

```python
# Создадим столбец 'категория_заказа_количество'
df0['категория_заказа_цена'] = df0.apply(price_cat, axis = 1)
```
